```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sqlite3 as sql
```

```python
df_data = pd.read_csv(r"portuguese-poems.csv")
```

```python
conn = sql.connect('DW.db')
cursor = conn.cursor()
```

```python
cursor.execute("""

CREATE TABLE Author (
        id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
        AuthorNome VARCHAR(100) NOT NULL

);
""")
```

```python
cursor.execute("""

CREATE TABLE Title (
        id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
        TitlePoema VARCHAR(500) NOT NULL

);
""")
```

```python
cursor.execute("""

CREATE TABLE Content (
        id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
        ContentPoema TEXT NOT NULL
);

""")
```

```python
cursor.execute("""

CREATE TABLE PoemaDW (
        Author_id INTEGER NOT NULL,
        Title_id INTEGER NOT NULL,
        Content_id INTEGER NOT NULL,
        Views INTEGER NOT NULL,
        Primary key(Author_id, Title_id, Content_id),
        Foreign key(Author_id) References Author(id) on delete cascade on update no action,
        Foreign key(Title_id) References Title(id) on delete cascade on update no action,
        Foreign key(Content_id) References Content(id) on delete cascade on update no action
    
);

""")
```

```python
from sqlalchemy import create_engine
```

```python
df_Author = df_data.Author.unique()
print(df_Author)
```

```python
engine = create_engine('sqlite:///DW.db', echo=False)
df_Author = pd.DataFrame(df_Author, columns=['AuthorNome'])
print(df_Author)

df_Author.to_sql('Author', con=engine, if_exists='append', index=False)
```

```python
df_Title = df_data.Title.unique()
print(df_Title)
```

```python
engine = create_engine('sqlite:///DW.db', echo=False)
df_Title = pd.DataFrame(df_Title, columns=['TitlePoema'])
print(df_Title)

df_Title.to_sql('Title', con=engine, if_exists='append', index=False)
```

```python
df_Content = df_data.Content.unique()
print(df_Content)
```

```python
engine = create_engine('sqlite:///DW.db', echo=False)
df_Content = pd.DataFrame(df_Content, columns=['ContentPoema'])
df_Content = df_Content.dropna()

df_Content.to_sql('Content', con=engine, if_exists='append', index=False)
```

```python
df_data = df_data.dropna()

for index, row in df_data.iterrows():
    #cursor.execute("select id from Author where id = 2").fetchall()
    
    id_Author = (cursor.execute("select id from Author where AuthorNome = '"+row["Author"].replace("'","\'\'")+"'").fetchone()[0])
    id_Title = (cursor.execute("select id from Title where TitlePoema = '"+row["Title"].replace("'","\'\'")+"'").fetchone()[0])
    id_Content = (cursor.execute("select id from Content where ContentPoema = '"+row["Content"].replace("'","\'\'")+"'").fetchone()[0])
    df_dwPoema = pd.DataFrame([[id_Author, id_Title, id_Content, row['Views']]], columns=['Author_id', 'Title_id', 'Content_id', 'Views'])
    existe = cursor.execute("select * from PoemaDW where Author_id = "+str(id_Author)+' and Title_id = '+str(id_Title)+' and Content_id = '+str(id_Content)).fetchone()
    if existe==None:
        df_dwPoema.to_sql('PoemaDW', con=engine, if_exists='append', index=False)
    print('ok')
```
