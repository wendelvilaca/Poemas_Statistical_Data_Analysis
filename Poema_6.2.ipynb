# 2 Classificação Ouro, Prata e Bronze

```python
#Perfil de visualização
df_RankingAutor = df_data.drop(columns=["Title","Content"])

interval = (0, int(Q2), int(Q3), np.inf)

ranking = ['Bronze','Prata', 'Ouro']

df_RankingAutorData = df_RankingAutor

df_RankingAutorData["RankingViews"] = pd.cut(df_RankingAutor.Views, interval, labels=ranking)

print(df_RankingAutorData)

#Visualizações por ranking Views

#df_SRankingViews = df_RankingViews.groupby(by="RankingViews").Views.sum()
#df_SRankingViewsI = df_SRankingViews.reset_index()
#df_SRankingViewsI = df_SRankingViewsI.sort_values(by="Views", ascending=False)

#print(df_SRankingViewsI)
#5 autores mais visualizados
#df_Top5Index = df_Top5Index.head(5)

#Plot
#df_SRankingViewsI.plot('RankingViews', 'Views', kind='bar')
#plt.show()
```

```python
df_RankingQuadroPre = df_RankingAutorData.drop(columns=["Views"])

df_RankingQuadroPre[['Ouro']] = df_RankingQuadroPre[['RankingViews']]
df_RankingQuadroPre[['Prata']] = df_RankingQuadroPre[['RankingViews']]
df_RankingQuadroPre[['Bronze']] = df_RankingQuadroPre[['RankingViews']]

df_RankingQuadroPro = df_RankingQuadroPre

Ouro={"Ouro" : "1","Prata" :"0","Bronze" : "0"}
df_RankingQuadroPro["Ouro"]=df_RankingQuadroPro["Ouro"].map(Ouro)

Prata={"Ouro" : "0","Prata" :"1","Bronze" : "0"}
df_RankingQuadroPro["Prata"]=df_RankingQuadroPro["Prata"].map(Prata)

Bronze={"Ouro" : "0","Prata" :"0","Bronze" : "1"}
df_RankingQuadroPro["Bronze"]=df_RankingQuadroPro["Bronze"].map(Bronze)

#df_RankingQuadroPro = pd.DataFrame(df_RankingQuadroPro['Author'].value_counts())
#df_RankingQuadroPro = df_RankingQuadroPro.groupby(by="Author").Ouro.count()
df_RankingQuadroPro = df_RankingQuadroPro.reset_index()

#print(df_RankingQuadroPro.groupby(by=['index','Author','Ouro','Prata','Bronze']).sum())

df_RankingQuadroProCol = df_RankingQuadroPro
#df_RankingQuadroProCol['Total'] = df_RankingQuadroPro.sum(axis=1)

df_RankingQuadroProCol['Ouro'] = df_RankingQuadroProCol['Ouro'].values.astype(np.int64)
df_RankingQuadroProCol['Prata'] = df_RankingQuadroProCol['Prata'].values.astype(np.int64)
df_RankingQuadroProCol['Bronze'] = df_RankingQuadroProCol['Bronze'].values.astype(np.int64)

df_RankingQuadroProCol['Total'] = df_RankingQuadroProCol['Ouro'] + df_RankingQuadroProCol['Prata'] + df_RankingQuadroProCol['Bronze']


df_RankingQuadroF = df_RankingQuadroProCol.groupby(by=['Author']).sum().groupby(level=[0]).cumsum()

df_RankingQuadroF = df_RankingQuadroF.sort_values(by='Ouro', ascending=False)

df_RankingQuadroF= df_RankingQuadroF.drop(columns=["index"])

print(df_RankingQuadroF.head(20))


#print(df_RankingQuadroProCol)
```

```python
#Perfil de visualização
df_Ranking = df_hist

interval = (0, int(Q2), int(Q3), np.inf)

ranking = ['Bronze','Prata', 'Ouro']

df_RankingViews = df_Ranking

df_RankingViews["RankingViews"] = pd.cut(df_Ranking.Views, interval, labels=ranking)
#print(df_RankingViews)

#Visualizações por ranking Views

df_SRankingViews = df_RankingViews.groupby(by="RankingViews").Views.sum()
df_SRankingViewsI = df_SRankingViews.reset_index()
df_SRankingViewsI = df_SRankingViewsI.sort_values(by="Views", ascending=False)

print(df_SRankingViewsI)
#5 autores mais visualizados
#df_Top5Index = df_Top5Index.head(5)

#Plot
df_SRankingViewsI.plot('RankingViews', 'Views', kind='bar')
plt.show()
```

```python
#Quantidade por ranking
dfa = pd.DataFrame(df_RankingViews['RankingViews'].value_counts())
dfa['%'] = 100*dfa['RankingViews']/df_RankingViews['Views'].shape[0]
dfa1 = dfa.reset_index()
print(dfa1)

#Plot RankingViews
dfa1.plot('index','RankingViews', kind='bar')
plt.show()
```
